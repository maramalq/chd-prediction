```python
# import libraries
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
%matplotlib inline
import pylab as plt
import seaborn as sns
```

```python
# loading data
df = pd.read_csv("cleaned_disease.csv")
df.head()
```

```python
# checking dataset structure
df.shape
```

### Continuous Features 

#### 1. sbp Feature

```python
# calculating summary statistics for sbp 
summary_statistics = df['sbp'].describe()
summary_statistics['mode'] = df['sbp'].mode()[0]
summary_statistics['variance'] = df['sbp'].var()
summary_statistics['skewness'] = df['sbp'].skew()
summary_statistics['range'] =  np.ptp(df['sbp'])
summary_statistics['sum'] = df['sbp'].sum()
print('Summary Statistics\n',summary_statistics)
```

```python
fig = plt.figure(figsize =(13, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

for s in ['top', 'bottom', 'left', 'right']:
    ax1.spines[s].set_visible(False)
    ax2.spines[s].set_visible(False)

ax1.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6)
n, bins, patches = ax1.hist(df['sbp'], bins=40, color='#167D7F')
ax1.set_xlabel('sbp')
ax1.set_ylabel('Frequency')

ax2.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6)
ax2 = sns.distplot(df['sbp'], hist=True, kde=True, bins=40, color='#167D7F', kde_kws={'linewidth': 1})
ax2.set_xlabel('sbp')
ax2.set_ylabel('Density')
plt.savefig('sbp.jpg')
plt.show()
```

### Categorical Features 

#### 9. Famhist Feature

```python
# calculating summary statistics 
df['famhist'].value_counts()
```

```python
fig = plt.figure(figsize =(13, 5))
ax1 = fig.add_subplot(1, 1, 1)
for s in ['top', 'bottom', 'left', 'right']:
    ax1.spines[s].set_visible(False)

ax1.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.7, alpha = 0.6)
ax1 = df['famhist'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0,color=['#167D7F','#5ab3b4'])
plt.xlabel("famhist", labelpad=10)
plt.ylabel("Count", labelpad=10)
plt.title("family history whether there is a prevalence of heart disease ", y=1.02);
plt.savefig('famhist.jpg')
```

```python
df['famhist'].value_counts().plot(kind='pie', figsize=(7, 6),colors=['#167D7F','#5ab3b4'],autopct='%1.1f%%')
#draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')  
plt.tight_layout()
plt.savefig('famhist2.jpg')
plt.show()
```
